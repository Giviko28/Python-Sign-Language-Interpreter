```python
filepath = "data/nox.csv"
df = pd.read_csv(filepath)
```

```python
df.index = df.index.astype(str)

# Manually filter the rows based on their index names
working_days = df[df['Unnamed: 0'].str.contains('working_day')]
non_working_days = df[df['Unnamed: 0'].str.contains('non_working_day')]

working_days = working_days.drop(columns=['Unnamed: 0'])
non_working_days = non_working_days.drop(columns=['Unnamed: 0'])

# Transpose to get hours as columns
working_days_T = working_days.T
non_working_days_T = non_working_days.T
```

```python
# Calculate the average NOx levels for each hour for working and non-working days
avg_working_days = working_days_T.mean(axis=1)
avg_non_working_days = non_working_days_T.mean(axis=1)

print(avg_working_days)

print(avg_non_working_days)
```

```python
# Calculate the average NOx level per day
avg_nox_working = working_days.mean(axis=0).mean()
avg_nox_non_working = non_working_days.mean(axis=0).mean()

print(f"\nAverage NOx level on working days: {avg_nox_working}")
print(f"Average NOx level on non-working days: {avg_nox_non_working}")
```

```python
# Finding the peak hour
peak_hour_working = avg_working_days.idxmax()
peak_hour_non_working = avg_non_working_days.idxmax()

print(f"\nPeak hour for NOx levels on working days: {peak_hour_working}")
print(f"Peak hour for NOx levels on non-working days: {peak_hour_non_working}")
```

```python
std_working_days = working_days_T.std(axis=1)
std_non_working_days = non_working_days_T.std(axis=1)

print(std_working_days)

print(std_non_working_days)
```

```python
# Hours of the day with the lowest NOx levels
min_hour_working = avg_working_days.idxmin()
min_hour_non_working = avg_non_working_days.idxmin()

print(f"\nHour with lowest NOx levels on working days: {min_hour_working}")
print(f"Hour with lowest NOx levels on non-working days: {min_hour_non_working}")
```

```python
# Combine the DataFrames
working_days['label'] = 1
non_working_days['label'] = 0  
combined_df = pd.concat([working_days, non_working_days])

X = combined_df.drop('label', axis=1).values
y = combined_df['label'].values

model = RandomForestClassifier(random_state=42)

cv_scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')

print(f"\nCross-validation scores: {cv_scores}")
print(f"Mean cross-validation score: {cv_scores.mean()}")

model.fit(X, y)
```
